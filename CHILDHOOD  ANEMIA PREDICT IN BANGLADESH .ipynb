### Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

lr=LogisticRegression()
lr.fit(x_train,y_train)

predlr=lr.predict(x_test)

print(f" Train Accuracy : {lr.score(x_train,y_train)}")
print(f" Test Accuracy : {lr.score(x_test,y_test)}")

#print(confusion_matrix(y_test, predlr))
#print(classification_report(y_test, predlr))
#plot_confusion_matrix(lr, x_test, y_test)  
#plt.show()
```

Sensitivity,Specificity,Cohen Kappa

```python
from sklearn.metrics import cohen_kappa_score
cmSVC = confusion_matrix(y_test,lr.predict(x_test))

#If we use TP TN FP and FN of below's comment we get range
#FP = cmSVC.sum(axis=0) - np.diag(cmSVC)  
#FN = cmSVC.sum(axis=1) - np.diag(cmSVC)
#TP = np.diag(cmSVC)
#TN = cmSVC.sum() - (FP + FN + TP)

TP = cmSVC[1,1]  
TN = cmSVC[0,0] 
FP = cmSVC[0,1] 
FN = cmSVC[1,0] 

# Sensitivity, hit rate, recall, or true positive rate
TPR = TP/float(TP+FN)
# Specificity or true negative rate
TNR = TN/float(TN+FP) 

# Cohen Kappa
cohen_score = cohen_kappa_score(y_test, predlr)

print("Sensitivity/TPR: " + str(TPR))
print("Specificity/TNR: " + str(TNR))
print("Cohen Kappa: " + str(cohen_score))
```

AUC Score

```python
from sklearn.metrics import roc_auc_score

# predict probabilities
pred_prob1 = lr.predict_proba(x_test)
#pred_prob2 = lr.predict_proba(x_test)
#pred_prob3 = lr.predict_proba(x_test)
print(f" Train AUC : {lr.roc_auc_score(x_train,y_train)}")
print(f" Test AUC : {lr.roc_auc_score(x_test,y_test)}")
# auc scores
auc_score1 = roc_auc_score(y_test, pred_prob1,multi_class='ovr')
#auc_score2 = roc_auc_score(y_test, pred_prob2[:,1],multi_class='ovr')
#auc_score3 = roc_auc_score(y_test, pred_prob3[:,1],multi_class='ovr')
print(auc_score1)
```
