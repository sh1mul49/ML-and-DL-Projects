```python
# Resize 
# Method-1: Specifying scaling factor using fx and fy
r1=cv2.resize(img,None,fx=10,fy=10)
plt.imshow(r1[:,:,::-1])
```

```python
# Resize method 2: Specify exact size of the image
#d_w=300,
#d_h=500
#dim=(d_w,d_h)
r2=cv2.resize(img,dsize=(300,500),)
plt.imshow(r2[:,:,::-1])
```

```python
# Resize method-3: Resize by maintaining aspect ratio
d_w=200
asp_r=d_w/img.shape[1]# height of the original image
d_h=int(img.shape[0]*asp_r)
r3=cv2.resize(img,dsize=(d_w,d_h))
plt.imshow(r3[:,:,::-1])
```

```python
# Flip image
fx=cv2.flip(img,0)# flip aroud x asis
fy=cv2.flip(img,1)#flip around y asis
fb=cv2.flip(img,-1)# flip around both axis
plt.figure(figsize=[20,10])
plt.subplot(141);plt.imshow(img[:,:,::-1]);plt.title('Original');
plt.subplot(142);plt.imshow(fx[:,:,::-1]);plt.title('Horizontal');
plt.subplot(143);plt.imshow(fy[:,:,::-1]);plt.title('Vertical');
plt.subplot(144);plt.imshow(fb[:,:,::-1]);plt.title('Both flipped');
```

Image Enhancement

```python
#  increase/ decrease brigtness
matrix=np.ones(img.shape,dtype='uint8')* 100;
img_b=cv2.add(img,matrix)
img_d=cv2.subtract(img,matrix)
plt.figure(figsize=[20,10])
plt.subplot(131);plt.imshow(img[:,:,::-1]),plt.title('Original')
plt.subplot(132);plt.imshow(img_b[:,:,::-1]),plt.title('Brighter')
plt.subplot(133);plt.imshow(img_d[:,:,::-1]),plt.title('Darker')
```

```python
# Multiplication to increase/decrease Contrast

matrix1=np.ones(img.shape)* 1.2
matrix2=np.ones(img.shape)* .4
img_b=np.uint8(cv2.multiply(np.float64(img),matrix1))
img_d=np.uint8(cv2.multiply(np.float64(img),matrix2))
plt.figure(figsize=[20,10])
plt.subplot(131);plt.imshow(img[:,:,::-1]),plt.title('Original')
plt.subplot(132);plt.imshow(img_b[:,:,::-1]),plt.title('High contrast')
plt.subplot(133);plt.imshow(img_d[:,:,::-1]),plt.title('Low contrast')
```
