```python
# importing the required modules  
from chatterbot import ChatBot  
from chatterbot.trainers import ListTrainer  
from chatterbot.trainers import ChatterBotCorpusTrainer  
  
```

```python
# creating a chatbot  
HelpBot = ChatBot(  
    name = 'Shimul',  
    read_only = True,  
    logic_adapters = [  
        'chatterbot.logic.MathematicalEvaluation',  
        'chatterbot.logic.BestMatch'  
        ]  
        )  
```

```python
# using the ChatterBotCorpusTrainer class  
corpus_trainee = ChatterBotCorpusTrainer(HelpBot)  
corpus_trainee.train('/content/drive/MyDrive/Chatbot/Chatbot/data') 
```

```python
#@title Talk with HSTU
You="Where is the bank location"#@param {allow-input: true} {type:"string"}
print("HelpBot:",HelpBot.get_response(You))
```
